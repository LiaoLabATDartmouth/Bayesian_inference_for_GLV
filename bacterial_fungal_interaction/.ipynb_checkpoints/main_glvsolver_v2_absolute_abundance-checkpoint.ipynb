```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cmdstanpy
import sys
sys.path.append('../')
from GlvSolver_v2 import *
```

# Load MSK data

## read sample metadata (day -10 to 40)

```python
df_sample = pd.read_csv('MSK_Data/tblsamples.csv', index_col=0)
df_sample.PatientID = df_sample.PatientID.astype(str)
df_sample = df_sample[df_sample.DayRelativeToNearestHCT.notnull()]
df_sample = df_sample[df_sample.SampleType == 'Stool']
df_sample = df_sample[(df_sample.DayRelativeToNearestHCT >= -10) & (df_sample.DayRelativeToNearestHCT <= 40)]
df_sample.head()
```

## read bacterial composition

```python
df_bact_count = pd.read_csv("MSK_Data/tblcounts_melt_bact.csv")
df_bact_tax = pd.read_csv("MSK_Data/tblASVtaxonomy_silva138_bact.csv")
df_bact_count = pd.merge(df_bact_count, df_bact_tax[['ASV','Genus']], left_on='ASV', right_on='ASV', how='left').fillna('<not present>')
df_bact_count = pd.pivot_table(df_bact_count, index="SampleID", columns="Genus", values="Count", aggfunc="sum").fillna(0)
df_bact_relab = df_bact_count.div(df_bact_count.sum(axis=1), axis=0)
df_bact_relab = df_bact_relab.drop('<not present>', axis=1)

# multiply by 16S qPCR
df_16s = pd.read_csv("MSK_Data/tblqpcr16S.csv")
df_16s = df_16s.groupby('SampleID')['qPCR16S'].agg('mean').reset_index().set_index('SampleID')
df_16s = df_16s[df_16s.qPCR16S > 0]
df_bact_relab = pd.merge(df_bact_relab, df_16s, left_index=True, right_index=True, how='inner')
for col in df_bact_relab.columns:
    if col != 'qPCR16S':
        df_bact_relab[col] *= df_bact_relab['qPCR16S']
df_bact_relab = df_bact_relab.drop('qPCR16S', axis=1)

df_bact_relab.head()
```

## retrieve C. parapsilosis complex relative abundance

```python
df_fungi_count = pd.read_csv("MSK_Data/tblcounts_melt_fungi.csv")
df_fungi_tax = pd.read_csv("MSK_Data/tblASVtaxonomy_refseq_fungi_TR.csv")
df_fungi_count = pd.merge(df_fungi_count, df_fungi_tax[['ASV','Species']], left_on='ASV', right_on='ASV', how='left').fillna('<not present>')
df_fungi_count = pd.pivot_table(df_fungi_count, index="SampleID", columns="Species", values="Count", aggfunc="sum").fillna(0)
df_fungi_relab = df_fungi_count.div(df_fungi_count.sum(axis=1), axis=0)
df_fungi_relab = df_fungi_relab[['Candida parapsilosis', 'Candida metapsilosis', 'Candida orthopsilosis']]
df_fungi_relab['Candida_parapsilosis_complex'] = df_fungi_relab.sum(axis=1)
df_fungi_relab = df_fungi_relab[['Candida_parapsilosis_complex']]

# multiply by 18S qPCR
df_18s = pd.read_csv("MSK_Data/tblqpcr18S.csv")
df_18s = df_18s.groupby('SampleID')['qPCR18S'].agg('mean').reset_index().set_index('SampleID')
df_18s = df_18s[df_18s.qPCR18S > 0]
df_fungi_relab = pd.merge(df_fungi_relab, df_18s, left_index=True, right_index=True, how='inner')
df_fungi_relab['Candida_parapsilosis_complex'] *= df_fungi_relab['qPCR18S']
df_fungi_relab = df_fungi_relab.drop('qPCR18S', axis=1)

df_fungi_relab.head()
```

## Update df_sample, df_bact_relab, and df_fungi_relab using shared samples

```python
shared_samples = list(set(df_sample.index).intersection(set(df_bact_relab.index)).intersection(set(df_fungi_relab.index)))
len(shared_samples)
```

```python
df_sample = df_sample.loc[shared_samples]
df_bact_relab = df_bact_relab.loc[shared_samples]
df_fungi_relab = df_fungi_relab.loc[shared_samples]
```

## Remove patients that have too few samples

```python
value_counts = df_sample['PatientID'].value_counts()
df_sample = df_sample[df_sample['PatientID'].isin(value_counts[value_counts >= 5].index)]
df_bact_relab = df_bact_relab.loc[df_sample.index]
df_fungi_relab = df_fungi_relab.loc[df_sample.index]
```

## Take top 20 bacterial genera and merge it with df_fungi_relab

```python
genus_total_abundance = df_bact_relab.sum(axis=0)
top_genera = genus_total_abundance.nlargest(20).index
df_bact_relab = df_bact_relab[top_genera]
df_bact_relab.head()
```

```python
df_bact_cpara = pd.merge(df_bact_relab, df_fungi_relab, left_index=True, right_index=True, how='inner')
df_bact_cpara.columns = ['Taxon_' + col for col in df_bact_cpara.columns]
df_bact_cpara = df_bact_cpara.rename(
    {'Taxon_Escherichia-Shigella':'Taxon_Escherichia_Shigella',
     'Taxon_[Clostridium] innocuum group':'Taxon_Clostridium_innocuum_group',
     'Taxon_[Ruminococcus] gnavus group':'Taxon_Ruminococcus_gnavus_group',
     'Taxon_[Ruminococcus] torques group':'Taxon_Ruminococcus_torques_group'
    },
    axis=1
)
df_bact_cpara.head()
```

```python
assert len(df_bact_cpara) == len(df_sample)
df_sample.head()
```

```python
len(df_bact_cpara), len(df_sample)
```

## Read antifungal drug information

```python
df_drug = pd.read_csv('MSK_Data/tbldrug.csv', low_memory=False)
df_drug.PatientID = df_drug.PatientID.astype(str)
df_drug = df_drug[df_drug.Category == 'antifungals']

# convert to single day entry
df_drug_single_day = []
for index in df_drug.index:
    start_tps = df_drug.loc[index,'StartTimepoint']
    stop_tps = df_drug.loc[index,'StopTimepoint']
    start_day = df_drug.loc[index,'StartDayRelativeToNearestHCT']
    for timepoint in np.arange(start_tps,stop_tps+1):
        res = list(df_drug.loc[index])
        res.append(timepoint)
        res.append(timepoint-start_tps+start_day)
        df_drug_single_day.append(res)
df_drug_single_day = pd.DataFrame(df_drug_single_day, columns=list(df_drug.columns)+['Timepoint','DayRelativeToNearestHCT'])
df_drug_single_day = df_drug_single_day.drop(['StartTimepoint','StopTimepoint','StartDayRelativeToNearestHCT','StopDayRelativeToNearestHCT'], axis=1)
df_drug = deepcopy(df_drug_single_day)
df_drug = df_drug[df_drug.DayRelativeToNearestHCT.notnull()]
df_drug = df_drug.drop_duplicates()
df_drug = df_drug[['PatientID','Factor','Timepoint']]

df_drug['Value'] = 1
df_drug = pd.pivot_table(df_drug, index=['PatientID','Timepoint'], columns='Factor', values='Value', aggfunc="mean").fillna(0).reset_index()

# add Perturbation as prefix
drug_variables = list(df_drug.columns[2:])
df_drug.columns = ['Perturbation_'+col if col in drug_variables else col for col in df_drug.columns]
df_drug = df_drug.rename({'Perturbation_isavuconazonium sulfate':'Perturbation_isavuconazole'}, axis=1)

# keep only micafungin
df_drug = df_drug.drop_duplicates()
df_drug.head()
```

# Generate the input table to GlvSolver
The input table has the following columns: subject id, sample id, time point, pertubation id 1, perturbation id 2, ..., perturbation id n, followed by taxon id 1, taxon id 2, ..., taxon id n. Perturbation ids must start with "Perturbation_", and Taxon ids must start with "Taxon_".

```python
# merge sample metadata and antifungal drug
df_input = pd.merge(
    df_sample.reset_index()[['PatientID','SampleID','Timepoint']],
    df_drug,
    left_on=['PatientID','Timepoint'],
    right_on=['PatientID','Timepoint'],
    how='left'
).fillna(0.0)
df_input = df_input.loc[:, (df_input != 0).any(axis=0)] # remove columns that are only zero

# merge further with bacterial/fungal taxa
df_input = pd.merge(
    df_input,
    df_bact_cpara,
    left_on='SampleID',
    right_on='SampleID',
    how='inner'
)

# change PatientID to SubjectID
df_input = df_input.rename({'PatientID':'SubjectID'}, axis=1)
df_input =  df_input.sort_values(['SubjectID','Timepoint'], ascending=True)

df_input.head()
```

```python
# make sure how NA in the table
assert df_input.isna().any().any() == 0
```

# Run GlvSolver and generate the input files to CMDStan

```python
df_output, scaling_factor = compute_dlogy_dt(df_input, method='spline')
df_output.head()
```

```python
Xmat, Ymat, sample_ids, taxa_ids, perturbation_ids = generate_XY_matrics(df_output)
```

```python
# only consider impact of antifungal drugs on C. parapsislosis
pairs_to_exclude = []
for tid in taxa_ids:
    if tid != 'Taxon_Candida_parapsilosis_complex':
        for pid in perturbation_ids:
            pairs_to_exclude.append((tid, pid))
```

```python
write_stan_input_file(
    prefix='BF_interact',
    stan_path_dir='glvsolver_v2_absolute_abundance/',
    Xmat=Xmat,
    Ymat=Ymat,
    taxa_ids=taxa_ids,
    perturbation_ids=perturbation_ids,
    pairs_to_exclude=pairs_to_exclude,
    sigma_ub=1.0,
    normal_prior_std=1.0,
    neg_self_int=True
)
```

# Run CMDStan (go to glvsolver_v2_absolute_abundance folder and run "./run_cmdstan.sh")

# Parse CMDStan outputs

```python
stan_output_path = ["glvsolver_v2_absolute_abundance/BF_interact_%d.csv"%(i) for i in np.arange(1,4)]
df_parsed = parse_stan_output(
    stan_output_path=stan_output_path,
    taxa_ids=taxa_ids,
    perturbation_ids=perturbation_ids,
    pairs_to_exclude=pairs_to_exclude,
    bci_cutoff=0.90,
    scaling_factor=1,
    sig_only=False
).sort_values('Variable').set_index("Variable")
```

```python
# find interactions related to C. parapsilosis
df_cpara_interactions = df_parsed[(df_parsed.Sig==True) & (df_parsed.Taxon1.str.contains('Candida'))].sort_values('Mean')
df_cpara_interactions.to_csv("glvsolver_v2_absolute_abundance/cpara_interactions_summary.csv")
df_cpara_interactions
```

```python
fit = cmdstanpy.from_csv(stan_output_path)
idata = az.from_cmdstanpy(posterior=fit)

# get C. parapsilosis index
cpara_idx = taxa_ids.index('Taxon_Candida_parapsilosis_complex')

# sort beta
ordered_beta_indices = np.argsort(np.mean(idata.posterior["constrained_beta"][:,:,cpara_idx,:].to_numpy(), axis=(0,1)))

# sort epsilon
ordered_epsilon_indices = np.argsort(np.mean(idata.posterior["epsilon"][:,:,cpara_idx,:].to_numpy(), axis=(0,1)))

# rename variables
beta_coord_labels = {}
for idx, taxon in enumerate(taxa_ids):
    beta_coord_labels[idx] = taxon

epsilon_coord_labels = {}
for idx, perturbation in enumerate(perturbation_ids):
    epsilon_coord_labels[idx] = perturbation

labeller = az.labels.MapLabeller(
    coord_map={
        'constrained_beta_dim_1': beta_coord_labels,
        'epsilon_dim_1': epsilon_coord_labels,
    }
)

fig, ax = plt.subplots(figsize=(3,5))
_ = ax.axvline(x=0, color='k', linestyle='--', linewidth=1)
_ = az.plot_forest(
    idata,
    var_names=["epsilon","constrained_beta","alpha"],
    coords={
        "alpha_dim_0": [cpara_idx],
        "constrained_beta_dim_0": [cpara_idx],
        "constrained_beta_dim_1": ordered_beta_indices,
        "epsilon_dim_1": ordered_epsilon_indices,
        "epsilon_dim_0": [cpara_idx]
    },
    linewidth=1,
    hdi_prob=0.90,
    textsize=10,
    markersize=10,
    ax=ax,
    combined=True,
    labeller=labeller
)

_ = ax.set_xlim([-4,4])

plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('glvsolver_v2_absolute_abundance/bacterial_fungal_interactions.svg', dpi=600, bbox_inches='tight')
```
